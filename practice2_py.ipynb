Student information system


```python
def details():
    name = input("Enter your name: ")
    print(f"The student name is: {name}")

    std_id = int(input("Enter your student id: "))
    print(f"The student id is: {std_id}")

    marks = int(input("Enter your marks: "))
    print(f"The student marks is: {marks}")


def sub():
    cnt = int(input("Enter the number of subjects: "))
    print(f"The student has {cnt} subjects")

    subject = []

    while cnt != 0:
        lst = input("Enter your subject name:\n")
        subject.append(lst)
        cnt -= 1

    print("Subjects are:", subject)


def subdetails():
    m = {}

    count = int(input("Enter number of subjects whose marks will be added in result: "))

    for i in range(count):
        key = input(f"Enter subject {i+1}: ")
        value = int(input(f"Enter marks for subject {i+1}: "))

        m[key] = value

    return m

def marks(data):
  total=0
  for i in data.values():
     total+=i


  print(total)



print("Welcome to the student marks analyzer system")
details()
sub()
a=subdetails()
marks(a)
```

Calculator function

```python
def calculator():
   n=int(input("Enter your choice:\n1.Addition\n2.Subtraction\n3.Multiplication\n4.Division\n"))
   if(n==1):
      print("Addition")
      a=int(input("Enter first number: "))
      b=int(input("Enter second number: "))
      print(a+b)

   elif(n==2):
      print("Subtraction")
      a=int(input("Enter first number: "))
      b=int(input("Enter second number: "))
      print(a-b)

   elif(n==3):
      print("Multiplication")
      a=int(input("Enter first number: "))
      b=int(input("Enter second number: "))
      print(a*b)

   elif(n==4):
      print("Division")
      a=int(input("Enter first number: "))
      b=int(input("Enter second number: "))
      print(a/b)

   else:
      print("Invalid choice")

calculator()
```
