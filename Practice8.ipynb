```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
import seaborn as sns
print("All libraries imported successfully")
```

```python
data={
    "Age":[21,20,19,20,18,20,21],
    "salary":[10000,12000,13000,11000,15000,13000,9000],
    "Spending": [20,25,30,45,50,55,60]
}
df=pd.DataFrame(data)
print(df)
```

```python
plt.scatter(df["salary"],df["Spending"])
plt.xlabel("salary")
plt.ylabel("Spending")
plt.title("Money use data")
plt.show()
```

Unsupervised ml

```python
X=df[["salary","Spending"]]
print(X)
```

```python
from sklearn.cluster import KMeans
kmodel=KMeans(n_clusters=3)
kmodel.fit(X)
print("Training sucessfull")
```

```python
prediction=kmodel.predict(X)
print(prediction)
```

```python
df["Clusters"]=prediction
print(df)
```

```python
plt.scatter(df["salary"],df["Spending"],c=df["Clusters"])
plt.xlabel("salary")
plt.ylabel("Spending")
plt.title("Money use data")
plt.show()
```

Centroid

```python
print(kmodel.cluster_centers_)
```

Inertia value(Elbow method)

```python
inertia_values = []

for k in range(1, 6):
    model = KMeans(n_clusters=k)
    model.fit(X)
    inertia_values.append(model.inertia_)

print(inertia_values)
```

Elbow graph

```python
plt.plot(range(1, 6), inertia_values)

plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")

plt.show()
```

PCA

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
print(X_scaled)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
model=pca.fit_transform(X_scaled)
print(model)
```

```python
plt.scatter(model[:,0], model[:,1])

plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA Visualization")

plt.show()
```
